# Romania shape 

Plotting th shape of Romania with tiles of 100, 10 and 1 km. 

Note that while Romania's shape resembles that of a fish, with head to the left and tail to the right, the tail is very alongated that the real map, as it is showing also the part of the Black Sea that Romania can exploit economically (oil, natural gas, etc).

data from EU: https://www.eea.europa.eu/data-and-maps/data/eea-reference-grids-2/gis-files/romania-shapefile

Then add the natural reserves from Romania.

data from EU: http://www.mmediu.ro/articol/date-gis/434

```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
INPUT_FOLDER_NAME = "/Users/abuzatu/Work/data/geo_spatial/Romania"
CRS_INIT = {"init": "epsg:4326"} # longitude - latitude coordinate system
```

# Romania map

```python
gdf_100 = gpd.read_file(f"{INPUT_FOLDER_NAME}/1/ro_100km.shp")
gdf_100
```

```python
gdf_100.plot()
```

```python
gdf_10 = gpd.read_file(f"{INPUT_FOLDER_NAME}/1/ro_10km.shp")
gdf_10.plot()
```

```python
gdf_1 = gpd.read_file(f"{INPUT_FOLDER_NAME}/1/ro_1km.shp")
gdf_1.plot()
```

# Add natural reserves

```python
gdf_reserves = gpd.read_file(f"{INPUT_FOLDER_NAME}/2/Limite_N2k_2017.shp")
gdf_reserves
```

```python
gdf_reserves.plot()
```

# Let's overlay the two together

```python
fig, ax = plt.subplots(1, 1, figsize = (10, 6))
gdf_10.plot(ax = ax, color = "cyan")
gdf_reserves.plot(ax = ax, color = "green")
```

They are at different scales, so let's convert first both gdf to the longitude and latitude coordinate system, which is CRS_INIT = {"init": "epsg:4326"}

```python
fig, ax = plt.subplots(1, 1, figsize = (10, 6))
gdf_10.to_crs(CRS_INIT).plot(ax = ax, color = "cyan")
gdf_reserves.to_crs(CRS_INIT).plot(ax = ax, color = "green")
plt.xlabel("Longitude", fontsize = 18)
plt.ylabel("Latitude", fontsize = 18)
plt.title("Natural reserves in Romania", fontsize = 18)
```

We note that the southern reserves are on the river Delta, which is also the frontier. Same the eastern ones on the river Prut, also the frontier. So we see a bit also the cyan across the boarders, as we use a grid of 10 km. Let's plt again with a grid of 1 km. 

```python
fig, ax = plt.subplots(1, 1, figsize = (10, 6))
gdf_1.to_crs(CRS_INIT).plot(ax = ax, color = "cyan")
gdf_reserves.to_crs(CRS_INIT).plot(ax = ax, color = "green")
plt.xlabel("Longitude", fontsize = 18)
plt.ylabel("Latitude", fontsize = 18)
plt.title("Natural reserves in Romania", fontsize = 18)
```

Unfortunately it's still showing things outside the frontier in cyan
